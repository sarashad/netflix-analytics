```python
# ببین کدوم سطرها هم director هم cast خالی دارن
both_null = df[df['director'].isnull() & df['cast'].isnull()]
print("Rows with both director AND cast missing:", len(both_null))

# ببین این سطرها بیشتر از کدوم type هستن
print("\nType breakdown:")
print(both_null['type'].value_counts())
```

```python
# کشورهایی که بیشترین director خالی دارن
print("Top countries with missing director:")
print(df[df['director'].isnull()]['country'].value_counts().head(10))
```

```python
# تعداد کل فیلم/سریال هر کشور
print("Total content by country:")
print(df['country'].value_counts().head(10))
```
