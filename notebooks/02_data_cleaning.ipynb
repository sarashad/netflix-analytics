```python
import pandas as pd 

df = pd.read_csv('../data/netflix_titles.csv') 
```

```python
# 1. Fill missing values
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown').str.split(', ')

# 2. Fix dates
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')

# 3. Fix duration
df.loc[df['type'] == 'Movie', 'duration_minutes'] = df.loc[df['type'] == 'Movie', 'duration'].str.split(' ').str[0].astype(float)
df.loc[df['type'] == 'TV Show', 'duration_seasons'] = df.loc[df['type'] == 'TV Show', 'duration'].str.split(' ').str[0].astype(float)

# 4. Drop rows with missing date_added and rating
df = df.dropna(subset=['date_added', 'rating'])

print("Done! Shape:", df.shape)
print(df['cast'].head(5))
```

```python
df.to_csv('../data/cleaned_data.csv', index=False) 
```
